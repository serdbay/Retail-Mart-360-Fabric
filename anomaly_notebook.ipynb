```python
# Welcome to your new notebook
# Type here in the cell editor to add code!
from pyspark.sql.functions import avg, stddev, col, count
from pyspark.sql.window import Window

window_country = Window.partitionBy("COUNTRY")

df_zscore = spark.read.table("silver_sales")\
    .withColumn("avg_sales", avg("SALES").over(window_country))\
    .withColumn("std_sales", stddev("SALES").over(window_country))\
    .withColumn("z_score", (col("SALES") - col("avg_sales")) / col("std_sales"))

df_anomali = df_zscore.filter(
    (col("z_score") > 3) | (col("z_score") < -3)
)

df_anomali.write.format("delta")\
    .mode("overwrite")\
    .saveAsTable("gold_sales_anomalies")

print(f"✅ gold_sales_anomalies güncellendi! Anomali sayısı: {df_anomali.count()}")
```
