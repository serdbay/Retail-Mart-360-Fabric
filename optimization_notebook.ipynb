```python
#optimization part
print("=" * 50)
print("⚡ DELTA LAKE OPTİMİZASYON BAŞLIYOR")
print("=" * 50)

# Optimize edilecek tablolar
tables = [
    "silver_sales",
    "dim_customer_scd2", 
    "gold_country_ranking",
    "gold_monthly_trend",
    "dim_geography_country",
    "gold_data_quality_log"
    #"pipeline_error_log",
]

# 1. OPTIMIZE — Küçük dosyaları birleştir
print("\n📌 ADIM 1: OPTIMIZE")
for table in tables:
    spark.sql(f"OPTIMIZE {table}")
    print(f"✅ {table} optimize edildi")

# 2. VACUUM — Eski dosyaları temizle
print("\n📌 ADIM 2: VACUUM")
for table in tables:
    spark.sql(f"VACUUM {table} RETAIN 168 HOURS")
    print(f"✅ {table} vacuum tamamlandı")

print("\n" + "=" * 50)
print("🎉 OPTİMİZASYON TAMAMLANDI!")
print("=" * 50)
```
